## Data Analysis and Visualization 



```python
# 1.	What is the distribution of customer types (e.g., Loyal vs. Disloyal)?

customer_type_counts = df['Customer Type'].value_counts()
print("Count of Customer Types:")
print(customer_type_counts) 
# Plot the distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=customer_type_counts.index, y=customer_type_counts.values, palette='viridis')
plt.title('Distribution of Customer Types', fontsize=16)
plt.xlabel('Customer Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10)
plt.show()
```

```python
# 2.	What is the age distribution of passengers?
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=30, kde=True, color='lightblue', alpha=0.7)
plt.title('Age Distribution of Passengers', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# 3.	How does gender influence satisfaction levels?

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender', hue='satisfaction', palette='coolwarm')
plt.title('Gender vs Satisfaction Levels', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Satisfaction', fontsize=10)
plt.show()
```

```python
gender_satisfaction_counts = df.groupby(['Gender', 'satisfaction']).size().reset_index(name='Count')

# Pivot the data for better readability
pivot_table = gender_satisfaction_counts.pivot(index='Gender', columns='satisfaction', values='Count').fillna(0)

# Calculate the percentages of satisfaction within each gender
pivot_table['Total'] = pivot_table.sum(axis=1)
pivot_table['Satisfied (%)'] = (pivot_table.get('satisfied', 0) / pivot_table['Total']) * 100
pivot_table['Dissatisfied (%)'] = (pivot_table.get('dissatisfied', 0) / pivot_table['Total']) * 100

# Display the result
print(pivot_table)
```

```python
# 4.	What is the relationship between flight class and satisfaction?
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Class', hue='satisfaction', palette='viridis')
plt.title('Flight Class vs Satisfaction Levels', fontsize=16)
plt.xlabel('Flight Class', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Satisfaction', fontsize=10)
plt.show()
```

```python
# 5.	Which service aspect has the highest average rating?

service_columns = [
    'Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking','Gate location', 'Food and drink', 'Online boarding'
]

service_avg_ratings = df[service_columns].mean().sort_values(ascending=False)

# Plot the average ratings
plt.figure(figsize=(8, 6))
sns.barplot(x=service_avg_ratings.index, y=service_avg_ratings.values, palette='viridis')
plt.title('Average Rating for Each Service Aspect', fontsize=16)
plt.xlabel('Service Aspect', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.show()

print(service_avg_ratings)
```

```python
# 6.	Is there a correlation between 'Inflight entertainment' and overall satisfaction?
print(df[['Inflight entertainment', 'satisfaction']].dtypes)

df['Inflight entertainment'] = pd.to_numeric(df['Inflight entertainment'], errors='coerce')
df['satisfaction'] = pd.to_numeric(df['satisfaction'], errors='coerce')

df['Inflight entertainment'] = df['Inflight entertainment'].fillna(df['Inflight entertainment'].median())
df['satisfaction'] = df['satisfaction'].fillna(df['satisfaction'].median())

correlation = df['Inflight entertainment'].corr(df['satisfaction'])

print(f"Pearson Correlation between 'Inflight entertainment' and 'satisfaction': {correlation:.2f}")
```

```python
# 7.	What is the average flight distance for satisfied vs. dissatisfied customers?

df['satisfaction'] = pd.to_numeric(df['satisfaction'], errors='coerce')
df['Flight Distance'] = pd.to_numeric(df['Flight Distance'], errors='coerce')

df['Flight Distance'] = df['Flight Distance'].fillna(df['Flight Distance'].median())
df['satisfaction'] = df['satisfaction'].fillna(df['satisfaction'].median())

df['Satisfaction Category'] = df['satisfaction'].apply(lambda x: 'Satisfied' if x == 1 else 'Dissatisfied')

avg_flight_distance = df.groupby('Satisfaction Category')['Flight Distance'].mean()

print("Average Flight Distance for Satisfied vs. Dissatisfied Customers:")
print(avg_flight_distance)

# Visualize the results using a bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_flight_distance.index, y=avg_flight_distance.values, palette='coolwarm')
plt.title('Average Flight Distance for Satisfied vs. Dissatisfied Customers', fontsize=16)
plt.xlabel('Satisfaction Category', fontsize=12)
plt.ylabel('Average Flight Distance (km)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

## Final Takeaways:
- Customer Loyalty: Understanding the customer types and their satisfaction helps airlines focus on retaining loyal customers and improving services for disloyal ones.

- Targeted Improvements: Based on age and gender differences in satisfaction, airlines can personalize services to cater to specific demographics.
- Flight Class: Enhancing services in lower classes like Economy could help increase overall satisfaction.
- Inflight Services: Ensuring entertainment and other comfort services are top-notch, especially in Economy, could lead to higher satisfaction.
- Longer Flights: Passengers on longer flights are more likely to value the quality of services provided, suggesting airlines should focus on improving the experience for long-haul passengers.
